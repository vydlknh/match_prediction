# Creating predictors

```python
matches["venue_code"] = matches["venue"].astype("category").cat.codes #string -> category -> num
```

```python
matches["opp_code"] = matches["opponent"].astype("category").cat.codes #string -> category -> num
```

```python
matches["hour"] = matches["time"].str.replace(":.+","",regex = True).astype("int")
```

```python
matches["day_code"] = matches["date"].dt.dayofweek
```

```python
matches["target"] = (matches["result"] == "W").astype("int")
```
