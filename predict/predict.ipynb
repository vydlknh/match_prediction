# Creating predictors

```python
matches["venue_code"] = matches["venue"].astype("category").cat.codes #string -> category -> num
```

```python
matches["day_code"] = matches["date"].dt.dayofweek
```

# Improving precision

```python
grouped_matches = matches.groupby("team")
```

```python
group = grouped_matches.get_group("Manchester City")
```

```python
group
```

```python
def rolling_avg(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed='left').mean() #left = ignore current week when creating avg
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols) #remove rows with missing values
    return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
rolling_avg(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_avg(x, cols, new_cols))
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('team')
```

```python
matches_rolling
```

```python
matches_rolling.index = range(matches_rolling.shape[0]) #unique value for each index
```

```python
matches_rolling
```

# Retraining ML model

```python
def make_preds(data,predictors):
    train = data[data["date"] < '2022-01-01']
    test = data[data["date"] > '2022-01-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual = test["target"], predicted = preds), index=test.index)
    precision = precision_score(test["target"], preds)
    return combined, precision
```

```python
combined, precision = make_preds(matches_rolling, predictors + new_cols)
```

# Combining predictions

```python
class MissingDict(dict):
    __missing__ = lambda self, key: key

map_values = {
    "Brighton and Hove Albion": "Brighton",
    "Manchester United": "Manchester Utd",
    "Newcastle United": "Newcastle Utd",
    "Tottenham Hotspur": "Tottenham",
    "West Ham United": "West Ham",
    "Wolverhampton Wanderers": "Wolves"
}
mapping = MissingDict(**map_values)
```

```python
mapping["West Ham United"]
```

```python
combined["new_team"] = combined["team"].map(mapping)
```

```python
merged = combined.merge(combined, left_on=["date", "new_team"], right_on=["date", "opponent"]) 
#merge dataframe onto itself, look for new team field & merge with oppoent field on other side
```
